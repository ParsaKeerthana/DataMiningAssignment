```python
from pycaret.regression import *
import pandas as pd

# Load the dataset into a DataFrame
df_wine = pd.read_csv("/content/WineQT.csv")  # Replace this with the path to your WineQT.csv file
```

```python
reg_setup = setup(df_wine, target='quality')
```

```python
compare_models()
```

```python
rf_model = create_model('rf')
```

```python
tuned_rf = tune_model(rf_model)
```

```python
plot_model(tuned_rf, plot='residuals')
```

```python
predictions = predict_model(tuned_rf, data=df_wine)
```

```python
final_rf = finalize_model(tuned_rf)
```
