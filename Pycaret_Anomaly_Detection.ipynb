```python
import pandas as pd

df_titanic = pd.read_csv("/content/titanic3.csv")
df_titanic.head()
```

```python
data = df_titanic.sample(frac=0.90, random_state=42)
data_unseen = df_titanic.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.anomaly import *

exp_ano101 = setup(data, normalize = True,
                   session_id = 1)
```

```python
model_forest = create_model('iforest')
```

```python
model_svm = create_model('svm')
```

```python
models()
```

```python
iforest_results = assign_model(model_forest)
iforest_results.head()
```

```python
plot_model(model_forest)
```

```python
predictions = predict_model(model_forest, data=data_unseen)
predictions.head()
```

```python
data_predictions = predict_model(model_forest, data = data)
data_predictions.head()
```
