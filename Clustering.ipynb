```python
import pandas as pd

df_iris = pd.read_csv("Iris.csv")
print(df_iris.head())
```

```python
features = df_iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia)
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
df_iris['Cluster'] = kmeans.fit_predict(features)
```

```python
import seaborn as sns

sns.scatterplot(data=df_iris, x='SepalLengthCm', y='SepalWidthCm', hue='Cluster', palette='Set1')
plt.show()
```
